### Topic Modelling NLTK

```python
data = []

with open('text.txt', "r") as f:
    for line in f:
        line = line.replace('\n','')
        if line == '':
            continue
        else:
            data.append(str(line))
            
f.close()
```

```python
data[0]
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
porter = PorterStemmer()
wordnet_lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = set(stopwords.words('english'))
```

```python
import string
exclude = set(string.punctuation)
```

```python
def clean_text(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punct_free = "".join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(wordnet_lemmatizer.lemmatize(word) for word in punct_free.split())
    return normalized
```

```python
l = [line.split() for line in data]
l[0]
```

```python
clean_text(data[0]).split()
```

```python
clean_doc = [clean_text(line).split() for line in data]
```

```python
clean_doc
```

```python
import gensim
from gensim import corpora

dictionary = corpora.Dictionary(clean_doc)
```

```python
### Corpora Matrix

doc_term_matrix = [dictionary.doc2bow(doc) for doc in clean_doc]
```

```python
lda = gensim.models.ldamodel.LdaModel
```

```python
ldamodel = lda(doc_term_matrix, num_topics=20, id2word=dictionary, passes=50)
```

```python
for idx, topic in ldamodel.print_topics(-1):
    print('Topic: {} Word: {}'.format(idx, topic))
```

```python
text = clean_doc[10]
print('Test, \n')
test_lda = dictionary.doc2bow(text)
```

```python
for index, score in sorted(ldamodel[test_lda], key=lambda tup: -1*tup[1]):
    print("Score: {}\t Topic: {}".format(score, ldamodel.print_topic(index, 5)))
```
